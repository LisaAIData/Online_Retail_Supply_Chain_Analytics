# 1. Groupby and aggregations

```python
''' Groupby
count() - Number of non-null observations
sum()   - Sum of values
mean()  - Mean of values
median()- Arithmetic median of values
min()   - Minimum
max()   - Maximum
mode()  - Mode
std()   - Standard deviation
var()   - Variance
'''
```

```python
retail = retail.reset_index()
```

```python
retail.head()
```

```python
retail.groupby('Country').head()
```

```python
##### Total sales for each country
retail.groupby('Country')['Quantity'].sum()
```

```python
### mean
retail.groupby(['Country','Description'])['Quantity','Price'].mean()
```

```python
#### Multiple aggregation
```

```python
a=retail.groupby(['Country','Description'])['Quantity','Price'].agg([np.mean,np.median])
a
```

```python
a = a.sort_index()
a.loc[('Australia','10 COLOUR SPACEBOY PEN'),('Price','median')]
```

```python
###### without index
## dropping level
```

```python
b=retail.groupby(['Country','Description'])['Quantity','Price'].agg([np.mean,np.median]).reset_index()
b
```

```python
b = b.droplevel(axis=1,level=0)
```

```python
b
```

```python
retail.groupby(['Country','Description']).agg(mean_qty=('Quantity',np.mean),
                                             mean_price=('Price',np.mean),
                                             median_qty=('Quantity',np.median),
                                             median_price=('Price',np.median)).reset_index()   
#有没有reset_index()都不一样，可创建不同类型表格
```

# Pivot Table 透视图

```python
country_date_qty = retail[['Country','date','Quantity','Price']]
country_date_qty
```

```python
country_pivoted = pd.pivot_table(country_date_qty,index = 'date',columns = 'Country', 
                                 values = 'Quantity',fill_value=0).reset_index()
country_pivoted
```

```python
country_pivoted[['date','United Kingdom']]
```

```python
country_pivoted = pd.pivot_table(country_date_qty,index = 'date',columns = 'Country', 
                                 values = ['Quantity','Price'],fill_value=0).reset_index()
country_pivoted
```

```python
country_pivoted = pd.pivot_table(country_date_qty,index = 'date',columns = 'Country', values = ['Quantity','Price'],
                                 fill_value=0,aggfunc = {'Quantity':np.sum, 'Price':np.mean}).reset_index()
```

```python
country_pivoted
```

# Join
三种方法：
     
     * Left join *

从右图表中，提取做图表变量对应的值。

值：若右侧有对应的值，则返回值；若右侧无对应值，返回NA。

变量名：只从右表中返回左表拥有的变量对应的值。若右表有'dada'变量，但左表'dada'变量，则不提取。

     * Iner join *
Only the common keys，左右两表同时拥有的key和values。

     * Outer join / Full join *
返回所有key和对应的values。

```python
import pandas as pd
desingnation_data = pd.DataFrame({'name':['Mike','Apple','Micro','Intel','Lisa','Google'],
                                 'title':['manager','supervisor','professor','associate','leader','sectionhead']})

age_data = pd.DataFrame({'name':['Mike','Apple','Lee','Intel','Lisa','Tom'],
                                 'age':['40','34','25','31','30','20']})
```

```python
desingnation_data
```

```python
age_data
```

```python
### left join
pd.merge(desingnation_data, age_data,how = 'left')
```

```python
# Inner join 选取交集共同项
pd.merge(desingnation_data, age_data, how = 'inner')
```

```python
## Outer join / Full join
df = pd.merge(desingnation_data, age_data, how = 'outer')
df
```

```python
### 将 NA 改为 0 
df = pd.merge(desingnation_data, age_data, how = 'outer')
df = df.fillna(0)
df
```
